```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import train_test_split

y = df.salary

features = ['date', 'rating', 'Remote', 'CA']

# Selecting relevant columns
X = df[features]
X.head()

# Train/validation split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# random forest model
random_forest_model = RandomForestRegressor(random_state=1)
random_forest_model.fit(train_X, train_y)
random_forest_predictions = random_forest_model.predict(val_X)
random_forest_mae = mean_absolute_error(random_forest_predictions, val_y)

print("Validation MAE: {:,.0f}".format(random_forest_mae))
```

Let's add features from company and location data created from encoded columns.

```python
y = df.salary

features2 = ['date', 'rating', 'Remote', 'CA', 'NY', 'TX', 'WA', 'NJ', 'IL','DC', 'GA', 'OR', 'MD','FL', 'MN', 'VA', 'ID', 
             'Amazon', 'TETCHTECH INC', 'workcog', 'Spacebar Solutions LLC', 'Talent Experts LLC', 'Underdog.io']

# Including encoded variables from influential states
X = df[features2]
X.head()

# Train/validation split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# random forest model
random_forest_model = RandomForestRegressor(random_state=1)
random_forest_model.fit(train_X, train_y)
random_forest_predictions = random_forest_model.predict(val_X)
random_forest_mae = mean_absolute_error(random_forest_predictions, val_y)

print("Validation MAE: {:,.0f}".format(random_forest_mae))
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(random_forest_model, random_state=1).fit(val_X, val_y)
eli5.show_weights(perm, feature_names = val_X.columns.tolist())
```

Let's compare the random forest result with the baseline linear regression model:

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression().fit(train_X, train_y)
linear_model_predictions = linear_model.predict(val_X)
linear_model_mae = mean_absolute_error(linear_model_predictions, val_y)
print(linear_model_mae)
```

```python
diff = linear_model_mae - random_forest_mae
percentage_change = diff/linear_model_mae
print(percentage_change)
```

After adding features for certain companies and states, the MAE has decreased, with a significant improvement over the baseline linear regression model. The model appears to be able to predict the close ballpark of what salary jobs will have. Gathering more data during the collection phase may have allowed the model to be more robust. 

From the permutation shuffle, date, rating, and whether or not a posting was remote were the most important features, followed by the location (Washington), company (Spacebar Solutions), and again location (California).
